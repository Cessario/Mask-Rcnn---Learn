```python
import numpy as np
import imutils
import time
import cv2
import os


weightsPath = os.path.sep.join(["mask-rcnn-coco/frozen_inference_graph.pb"])
configPath = os.path.sep.join(["mask-rcnn-coco/mask_rcnn_inception_v2_coco_2018_01_28.pbtxt"])


print("[INFO] loading Mask R-CNN from disk...")
net = cv2.dnn.readNetFromTensorflow(weightsPath, configPath)

labelsPath = os.path.sep.join(["mask-rcnn-coco/coco.names"])
LABELS = open(labelsPath).read().strip().split("\n")
print(LABELS)

np.random.seed(42)
COLORS = np.random.randint(0, 255, size=(len(LABELS), 3),dtype="uint8")
```

```python
vids = (["Video/p.mp4",
         "Video/p1.mp4",
         "Video/p2.mp4",
         "Video/p3.mp4",
         "Video/p4.mp4",
         "Video/p5.mp4",
         "Video/p6.mp4",
         "Video/p7.mp4",
         "Video/p8.mp4",
         "Video/p9.mp4",
         "Video/p10.mp4",
         "Video/p11.mp4",
         "Video/test_5.mp4"])

cap = cv2.VideoCapture(vids[3])

writer = None

mThresh = 0.3
conf = 0.5
personIdx = 0
color = (255,0,0)

while True:
    (grabbed, frame) = cap.read()
    
    if not grabbed:
        break

    blob = cv2.dnn.blobFromImage(frame, swapRB=True, crop=False)
    net.setInput(blob)
    
    start = time.time()
    (boxes, masks) = net.forward(["detection_out_final","detection_masks"])
    end = time.time()

    # loop detek object
    for i in range(0, boxes.shape[2]):

        classID = int(boxes[0, 0, i, 1])
        confidence = boxes[0, 0, i, 2]

        if classID == personIdx and confidence > conf:
            
            (H, W) = frame.shape[:2]
            box = boxes[0, 0, i, 3:7] * np.array([W, H, W, H])
            (startX, startY, endX, endY) = box.astype("int")
            
            boxW = endX - startX
            boxH = endY - startY
            
            cX =  int(startX + (boxW/ 2))
            cY =  int(startY + (boxH/ 2))

#             mask = masks[i, classID]
#             mask = cv2.resize(mask, (boxW, boxH),
#                 interpolation=cv2.INTER_NEAREST)
#             mask = (mask > mThresh)


#             roi = frame[startY:endY, startX:endX][mask]
#             color = COLORS[classID]
#             blended = ((0.4 * color) + (0.6 * roi)).astype("uint8")

#             frame[startY:endY, startX:endX][mask] = blended

#             color = [int(c) for c in color]
            cv2.rectangle(frame, (startX, startY), (endX, endY),color, 2)
            cv2.circle(frame, (cX, cY), 3, color, -1)
#             text = "{}: {:.4f}".format(LABELS[classID], confidence)
#             cv2.putText(frame, text, (startX, startY - 5),cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)


    if writer is None:
        fourcc = cv2.VideoWriter_fourcc(*"MJPG")
        writer = cv2.VideoWriter("outputV.avi", fourcc, 30,(frame.shape[1], frame.shape[0]), True)


    if writer is not None:
        writer.write(frame)
        
    cv2.imshow("frame",frame)
    key = cv2.waitKey(100) & 0xFF
    if key == ord("p"):
        print("pause")
        cv2.waitKey(-1) 
    elif key == ord("q"):
        print("closing")
        break

print("[INFO] cleaning up...")
cap.release()
cv2.destroyAllWindows()
```
